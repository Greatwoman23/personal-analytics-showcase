```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


```

```python
# Load the Titanic dataset
titanic_data = pd.read_csv("Titanic-Dataset.csv")

# Display the first few rows of the dataset
titanic_data.head()
```

```python
# Bar chart of passenger class distribution
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Age distribution of passengers
plt.hist(titanic_data['Age'].dropna(), bins=20, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# Scatter plot of fare vs. age
plt.scatter(x='Age', y='Fare', data=titanic_data, alpha=0.5, color='orange')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs. Age')
plt.show()
```

```python
# Pie chart of gender distribution
gender_counts = titanic_data['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution on Titanic')
plt.show()
```

```python
# Box plot of age by passenger class
sns.boxplot(x='Pclass', y='Age', data=titanic_data)
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
# Distribution of fare by passenger class and survival
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=titanic_data)
plt.title('Distribution of Fare by Passenger Class and Survival')
plt.show()
```

```python
# Survival rate by embarked port
sns.barplot(x='Embarked', y='Survived', data=titanic_data, order=['C', 'Q', 'S'])
plt.title('Survival Rate by Embarked Port')
plt.show()
```

```python
# Age distribution of survivors vs. non-survivors
plt.figure(figsize=(10, 6))

# Convert infinite values to NaN
titanic_data['Age'] = titanic_data['Age'].replace([np.inf, -np.inf], np.nan)

sns.kdeplot(titanic_data[titanic_data['Survived'] == 1]['Age'], label='Survived', fill=True)
sns.kdeplot(titanic_data[titanic_data['Survived'] == 0]['Age'], label='Not Survived', fill=True)

plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution of Survivors vs. Non-survivors')
plt.legend()
plt.show()
```

```python
# Count of passengers with siblings/spouses aboard
sns.countplot(x='SibSp', data=titanic_data)
plt.title('Count of Passengers with Siblings/Spouses Aboard')
plt.show()
```

```python
# Pair plot of numeric features
sns.pairplot(titanic_data[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived')
plt.suptitle('Pair Plot of Numeric Features', y=1.02)
plt.show()
```

```python
# Count of passengers by survival status
sns.countplot(x='Survived', data=titanic_data, palette='pastel')
plt.title('Count of Passengers by Survival Status')
plt.xlabel('Survival Status (0 = Not Survived, 1 = Survived)')
plt.show()
```

```python
# Histogram of passenger ages
plt.hist(titanic_data['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Passenger Ages')
plt.show()
```

```python
# Pie chart of passenger class distribution
class_counts = titanic_data['Pclass'].value_counts()
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightblue'])
plt.title('Passenger Class Distribution')
plt.show()
```
