# Summary Statistics

```python
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Display the first 5 rows of the dataset
print(titanic.head())

# Calculate Summary Statistics

# Average Age of Titanic Passengers
average_age = titanic["Age"].mean()
print("Average Age:", average_age)

# Median Age and Ticket Fare
median_age_fare = titanic[["Age", "Fare"]].median()
print("Median Age and Fare:")
print(median_age_fare)

# Descriptive Statistics for Age and Fare
desc_stats = titanic[["Age", "Fare"]].describe()
print("Descriptive Statistics:")
print(desc_stats)

# Aggregating Statistics for Age (min, max, median, skew) and Fare (min, max, median, mean)
agg_stats = titanic.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
print("Aggregating Statistics:")
print(agg_stats)

# Aggregating Statistics Grouped by Category

# Average Age for Male vs. Female Passengers
avg_age_gender = titanic[["Sex", "Age"]].groupby("Sex").mean()
print("Average Age for Male vs. Female Passengers:")
print(avg_age_gender)

# Mean Ticket Fare Price for Sex and Cabin Class Combinations
mean_fare_sex_class = titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
print("Mean Ticket Fare Price for Sex and Cabin Class Combinations:")
print(mean_fare_sex_class)

# Count Number of Records by Category

# Number of Passengers in Each Cabin Class
passengers_per_class = titanic["Pclass"].value_counts()
print("Number of Passengers in Each Cabin Class:")
print(passengers_per_class)
```

```python
# Load the Titanic data
titanic = pd.read_csv("Titanic-Dataset.csv")

# Filter passengers who survived
survived_passengers = titanic[titanic['Survived'] == 1]

# Filter passengers in Class 1
class_1_passengers = titanic[titanic['Pclass'] == 1]

# Filter female passengers who survived
female_survivors = titanic[(titanic['Sex'] == 'female') & (titanic['Survived'] == 1)]

# Print the results
print("Passengers who survived:")
print(survived_passengers.head())

print("\nPassengers in Class 1:")
print(class_1_passengers.head())

print("\nFemale passengers who survived:")
print(female_survivors.head())
```
